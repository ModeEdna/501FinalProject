```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
# get dfs for squads in each WC

# get all dfs for 2002 world cup
url = "https://en.wikipedia.org/wiki/2002_FIFA_World_Cup_squads"
soup = BeautifulSoup(requests.get(url).content, "html.parser")
dfs = []
for i in range(0,32):
    tbl = soup.select("table")[i]
    # remove the (aged XX) part:
    for td in tbl.select("td:nth-of-type(3)"):
        td.string = td.contents[-1].split("(")[0]
    df = pd.read_html(str(tbl))[0]
    dfs.append(df)
# concatenate the dfs into one
all2002 = pd.concat(dfs, axis=0)
all2002['Year'] = 2002
# export df
all2002.to_csv('/Users/modeedna/github-classroom/anly501/anly-501-project-ModeEdna/data/projectData/2002squads.csv')

# get all dfs for 2006 world cup
url = "https://en.wikipedia.org/wiki/2006_FIFA_World_Cup_squads"
soup = BeautifulSoup(requests.get(url).content, "html.parser")
dfs = []
for i in range(0,32):
    tbl = soup.select("table")[i]
    # remove the (aged XX) part:
    for td in tbl.select("td:nth-of-type(3)"):
        td.string = td.contents[-1].split("(")[0]
    df = pd.read_html(str(tbl))[0]
    dfs.append(df)   
# concatenate the dfs into one
all2006 = pd.concat(dfs, axis=0) 
all2006['Year'] = 2006
# export df
all2006.to_csv('/Users/modeedna/github-classroom/anly501/anly-501-project-ModeEdna/data/projectData/2006squads.csv')

# get all dfs for 2010 world cup
url = "https://en.wikipedia.org/wiki/2010_FIFA_World_Cup_squads"
soup = BeautifulSoup(requests.get(url).content, "html.parser")
dfs = []
for i in range(0,32):
    tbl = soup.select("table")[i]
    # remove the (aged XX) part:
    for td in tbl.select("td:nth-of-type(3)"):
        td.string = td.contents[-1].split("(")[0]
    df = pd.read_html(str(tbl))[0]
    dfs.append(df)
# concatenate the dfs into one
all2010 = pd.concat(dfs, axis=0)
all2010['Year'] = 2010
# export df
all2010.to_csv('/Users/modeedna/github-classroom/anly501/anly-501-project-ModeEdna/data/projectData/2010squads.csv')

# get all dfs for 2014 world cup
url = "https://en.wikipedia.org/wiki/2014_FIFA_World_Cup_squads"
soup = BeautifulSoup(requests.get(url).content, "html.parser")
dfs = []
for i in range(0,32):
    tbl = soup.select("table")[i]
    # remove the (aged XX) part:
    for td in tbl.select("td:nth-of-type(3)"):
        td.string = td.contents[-1].split("(")[0]
    df = pd.read_html(str(tbl))[0]
    dfs.append(df)
# concatenate the dfs into one
all2014 = pd.concat(dfs, axis=0)
all2014['Year'] = 2014
# export df
all2014.to_csv('/Users/modeedna/github-classroom/anly501/anly-501-project-ModeEdna/data/projectData/2014squads.csv')

# get all dfs for 2018 world cup
url = "https://en.wikipedia.org/wiki/2018_FIFA_World_Cup_squads"
soup = BeautifulSoup(requests.get(url).content, "html.parser")
dfs = []
for i in range(0,32):
    tbl = soup.select("table")[i]
    # remove the (aged XX) part:
    for td in tbl.select("td:nth-of-type(3)"):
        td.string = td.contents[-1].split("(")[0]
    df = pd.read_html(str(tbl))[0]
    dfs.append(df)
# concatenate the dfs into one
all2018 = pd.concat(dfs, axis=0)
all2018['Year'] = 2018
# export df
all2018.to_csv('/Users/modeedna/github-classroom/anly501/anly-501-project-ModeEdna/data/projectData/2018squads.csv')

# create a df with all squads from all world cups
allSquads = pd.concat([all2002, all2006, all2010, all2014, all2018], axis=0)
# export the dataframe
allSquads.to_csv('/Users/modeedna/github-classroom/anly501/anly-501-project-ModeEdna/data/projectData/allSquads.csv')
```
